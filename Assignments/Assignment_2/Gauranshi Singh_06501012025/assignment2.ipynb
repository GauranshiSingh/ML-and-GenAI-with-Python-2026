```python
n = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= n:
    factorial = factorial * i
    i = i + 1

print(f"Factorial of {n} = {factorial}")
```

```python
name = input("Enter student name: ")
roll = input("Enter roll number: ")

subjects = ["Maths", "Science", "English", "Hindi", "Social Science"]
marks = []

for subject in subjects:
    m = float(input(f"Enter marks in {subject} (out of 100): "))
    marks.append(m)

total = sum(marks)
percentage = (total / 500) * 100

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 40:
    grade = "D"
else:
    grade = "F"

print("\n" + "="*38)
print("       STUDENT RESULT CARD")
print("="*38)
print(f"Name        : {name}")
print(f"Roll No     : {roll}")
print("-"*38)

for i in range(len(subjects)):
    print(f"{subjects[i]:<16}: {marks[i]}")

print("-"*38)
print(f"Total       : {total} / 500")
print(f"Percentage  : {percentage:.2f}%")
print(f"Grade       : {grade}")
print("="*38)
```
